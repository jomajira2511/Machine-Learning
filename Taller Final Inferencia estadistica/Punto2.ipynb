```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

# Datos del cuadrado latino
data = {
    "Lote": [1]*5 + [2]*5 + [3]*5 + [4]*5 + [5]*5,
    "Operador": [1,2,3,4,5]*5,
    "Formula": [
        'A','B','C','D','E',
        'B','C','D','E','A',
        'C','D','E','A','B',
        'D','E','A','B','C',
        'E','A','B','C','D'
    ],
    "Respuesta": [
        1,5,6,1,1,
        8,1,5,2,11,
        7,13,1,2,4,
        1,6,1,2,3,
        3,5,5,4,6
    ]
}

df = pd.DataFrame(data)

# ANOVA
modelo = ols('Respuesta ~ C(Formula) + C(Lote) + C(Operador)', data=df).fit()
anova = sm.stats.anova_lm(modelo, typ=2)
print(anova)

# Visualización opcional
plt.figure(figsize=(8,5))
sns.boxplot(x='Formula', y='Respuesta', data=df, palette='Set3')
plt.title("Rapidez de combustión por Formulación")
plt.xlabel("Formulación")
plt.ylabel("Respuesta (rapidez de combustión)")
plt.grid(True)
plt.show()
```

Interpretación estadística

Valor-p = 0.8055 > 0.05 → No significativo.
Se puede concluir que no hay evidencia estadística para afirmar que las formulaciones tienen efectos diferentes sobre la rapidez de combustión.

Lote:
Valor-p = 0.5692 > 0.05 → No significativo.
Se puede concluir que no hay evidencia de que los lotes de materia prima afecten la respuesta.

Operador:
